```python
import re
import pickle

# Initialize tasks list
tasks = []
tasks_printed = False

# Function to add a task
def add_task(task):
    tasks.append(task)
    save_tasks()
    return f'Task "{task}" added.'

# Function to fetch tasks
def fetch_tasks():
    return tasks

# Function to remove a task
def remove_task(task):
    if task in tasks:
        tasks.remove(task)
        save_tasks()
        return f'Task "{task}" removed.'
    else:
        return f'Task "{task}" not found.'

# Function to print tasks
def print_tasks():
    global tasks_printed
    print("Chatbot: Your tasks:")
    for task in tasks:
        print(task)
    tasks_printed = True

# Function to save tasks to a file
def save_tasks():
    with open('tasks.pkl', 'wb') as file:
        pickle.dump(tasks, file)

# Function to load tasks from a file
def load_tasks():
    try:
        with open('tasks.pkl', 'rb') as file:
            return pickle.load(file)
    except FileNotFoundError:
        return []

# Process user input
def process_input(user_input):
    global tasks_printed

    # Define chatbot responses
    responses = {
        'hello': 'Hi there!',
        'how are you': 'I am a chatbot. I do not have feelings.',
        'tell me a joke': 'Why don\'t scientists trust atoms? Because they make up everything!',
        'add task (.*)': lambda task: add_task(task),
        'fetch tasks': fetch_tasks,
        'remove task (.*)': lambda task: remove_task(task),
        'quit': 'Goodbye!'
    }

    # Check for tasks to be printed
    if not tasks_printed:
        print_tasks()

    # Check for user input in responses
    for pattern, response in responses.items():
        match = re.match(pattern, user_input.lower())
        if match:
            if callable(response):
                return response(*match.groups())
            else:
                return f'Chatbot: {response}'

    return 'Chatbot: I did not understand that.'

# Load tasks from file
tasks = load_tasks()

# Main chat loop
while True:
    user_input = input("You: ")
    if user_input.lower() == "quit":
        print("Chatbot: Goodbye!")
        break

    response = process_input(user_input)
    print(response)
```
